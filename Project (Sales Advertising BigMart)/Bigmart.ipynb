## TASK-5

```python
# A
df_marketing["TotalAdvt"]=df_marketing["TV"] + df_marketing["radio"] + df_marketing["newspaper"]
```

#### Here we creates the new column with name TotalAdvt which sum of the three columns ( TV, radio and newspaper).

```python
# B
y = pd.DataFrame(df_marketing['sales'])
X = pd.DataFrame(df_marketing['TotalAdvt'])
simple_model = LinearRegression().fit(X, y)
print("score=",simple_model.score(X, y))
print("coeff=",simple_model.coef_)
print("intercept=",simple_model.intercept_ )
```

#### Here  we creates the simple_model which predicts the values of sales acoording to given input of the TotalAdvt.

```python
# C
m=0.04984724
c=4.11069214
x=50000
y=(m*x)+c
print(y)
```

#### Here we uses the formula Y = mX + C, where Y is predicted value, m is slope or coefficient and C is intercept.
#### Value of X is passed to the predict function of the model fitted above.

```python
# D 
simple_model.predict([[50000]])
```

## TASK-6

```python
# A
X = pd.DataFrame(df_marketing[['TV','radio','newspaper']])
y = pd.DataFrame(df_marketing['sales'])
multi_model = LinearRegression().fit(X, y)
print("score=",multi_model.score(X, y))
print("coeff=",multi_model.coef_)
print("intercept=",multi_model.intercept_ )
```

#### Here  we creates the multi_model which predicts the values of sales acoording to given input of the three different columns or forms of advertisements (TV, radio and newspaper).

```python
# B-(I)
m1=0.04704482;m2=0.18844353;m3=0.00217936
c=2.74120566
x1=90000;x2=3000;x3=45000
y=((m1*x1)+c)+((m2*x2)+c)+((m3*x3)+c)
print(y)
```

```python
# B-(II)
m1=0.04704482;m2=0.18844353;m3=0.00217936
c=2.74120566
x1=290000;x2=0;x3=80000
y=((m1*x1)+c)+((m2*x2)+c)+((m3*x3)+c)
print(y)
```
