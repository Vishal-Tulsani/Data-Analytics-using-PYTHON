```python
""" 1 """
import matplotlib.pyplot as plt
X = range(1, 50)
Y = [(2 + (3 * value)) for value in X]
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line.')
plt.show()
```

```python
""" 3 """
fig = plt.figure()
fig.subplots_adjust(bottom=0.050, left=0.020, top = 0.600, right=0.900)
x1 = [2, 3, 5, 6, 8]

y1 = [1, 5, 10, 18, 20]

x2 = [3, 4, 6, 7, 9]

y2 = [2, 6, 11, 20, 22]

gridsize = (3, 2)
fig = plt.figure(figsize=(12, 8))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2, rowspan=2)
ax2 = plt.subplot2grid(gridsize, (2, 0))
ax3 = plt.subplot2grid(gridsize, (2, 1))
ax1.plot(x1,y1,color='blue',marker="*")
ax1.plot(x2,y2,color='red',marker="o")
ax2.plot(x1,y1)
ax3.plot(x2,y2)
```

```python
""" 4 """
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, popularity, color='black')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Language\n")
plt.xticks(x_pos, x)
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.show()

```

```python
plt.barh(x_pos, popularity, color=['red', 'black', 'green', 'blue', 'yellow', 'cyan'])
plt.xlabel("Popularity")
plt.ylabel("Languages")
plt.title("PopularitY of Programming Language\n" )
plt.yticks(x_pos, x)
plt.show()
```

```python
""" 5 """
fig, ax = plt.subplots()
ax.add_patch(plt.Circle((0, 0),3, color='y'))
ax.add_patch(plt.Circle((9, 1), 4, hatch="/"))
ax.annotate("o", (0,0))
ax.annotate("(0,0)",(1,0))
ax.annotate("x", (9,1))
ax.annotate("(9,1)",(9,0))
ax.set_aspect('equal')
ax.plot()   
plt.show()
```
