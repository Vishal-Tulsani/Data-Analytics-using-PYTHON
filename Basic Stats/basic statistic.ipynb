## TASK-5

```python
# A
birth_data.rolling(12).mean()
```

```python
birth_data.rolling(12).std()
```

```python
# B
dynamic_mean = birth_data.rolling(12).mean()[:50]
dynamic_sd = birth_data.rolling(12).std()[:50]
plt.gcf().set_size_inches(15.7,12.27)
plt.gca().set(title="birth_data", xlabel="Date", ylabel="DailyCount")
plt.plot(birth_data[:50],'b',label='Orginal Data')
plt.plot(dynamic_mean,'r',label='Dynamic Mean')
plt.plot(dynamic_sd,'g',label='Dynamic SD')
plt.legend()
plt.xticks(rotation='vertical')
plt.show()
```

#### Here in this plot we uses only small data.
#### Here we plot the original data along with dynamic_mean (Dynamic Mean) and dynamic standard deviation (dynamic SD).
#### The plot shown has the three (3) different graphs.Here the BLUE one shows the original data, RED one tells us about Dynamic Mean and the GREEN one tells us about the Dynamic Standard Deviation.

### According to the plot the series is not stationary because here the value of Dynamic Mean as well as Dynamic Standard Deviation is varying with time.
### A stationary series is one in which the properties â€“ mean, variance and covariance, do not vary with time.

## TASK-6

```python
sm.graphics.tsa.plot_acf(birth_data)
plt.gcf().set_size_inches(115.7,112.27)
plt.show()
```

```python
sm.graphics.tsa.plot_pacf(birth_data)
plt.gcf().set_size_inches(15.7,12.27)
plt.show()
```

```python
ts_model = sm.tsa.ARMA(birth_data, (2,0)).fit(disp=False)
```

#### Here we use the ARMA model for making the series stationary.
####  ARMA model initially as cases of AR and MA model will be covered there. Since, ARMA model accepts 2 values p and q. If p =0 then the ARMA model is nothing but a simple MA series and for q = 0 it is simple a AR series. 

## TASK-7

```python
# A
print(ts_model.params)
```

```python
print(ts_model.aic, ts_model.bic, ts_model.hqic)
```

```python
# B
plt.plot(ts_model.resid,label='ts_model')
plt.plot(birth_data,label='original')
plt.legend()
plt.show()
```

#### Here we plot the graph between original data and the ts_model which has the ARMA model data.
#### Here you can see the difference after the ARMA model for making the data stationary.

## TASK-8

```python
# A
model_residual = ts_model.resid
```

```python
# B
model_residual.rolling(12).mean()
```

```python
model_residual.rolling(12).std()
```

```python
dynamic_mean = model_residual.rolling(12).mean()[:50]
dynamic_sd = model_residual.rolling(12).std()[:50]
plt.gcf().set_size_inches(15.7,12.27)
plt.gca().set(title="birth_data", xlabel="Date", ylabel="DailyCount")
plt.plot(birth_data[:50],'b',label='Model Residual')
plt.plot(dynamic_mean,'r',label='Dynamic Mean')
plt.plot(dynamic_sd,'g',label='Dynamic SD')
plt.legend()
plt.xticks(rotation='vertical')
plt.show()
```

#### Here in this plot we uses only small data.
#### Here we plot the original data along with dynamic_mean (Dynamic Mean) and dynamic standard deviation (dynamic SD).
#### The plot shown has the three (3) different graphs.Here the BLUE one shows the original data, RED one tells us about Dynamic Mean and the GREEN one tells us about the Dynamic Standard Deviation.
#### According to the plot the series is stationary because here the value of Dynamic Mean as well as Dynamic Standard Deviation is varying with time.
#### Here we convert it into the stationary from non stationary with the help of a ARMA model.
