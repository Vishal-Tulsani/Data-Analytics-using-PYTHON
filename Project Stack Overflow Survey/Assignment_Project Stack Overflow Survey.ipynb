```python
# D
df_sal_NonNa=t2
df_sal_NonNa.dropna()
```

```python
# E
sb.distplot(df_sal_NonNa.dropna(),bins=30,kde=False)
```

#### Here we draw the histogram graph with the help of a distplot() in which we enter the non NULL(Nan) of Converted Salary and uses the bins (the range of values—that is, divide the entire range of values into a series of intervals—and then count how many values fall into each interval) of size euals to 30. 

```python
sb.distplot(df_sal_NonNa.dropna(),bins=500,kde=False)
```

#### Here we draw the histogram graph with the help of a distplot() in which we enter the non NULL(Nan) of Converted Salary and uses the bins (the range of values—that is, divide the entire range of values into a series of intervals—and then count how many values fall into each interval) of size euals to 500. 

```python
# F
df_sal_low=df_sal_NonNa.dropna()[ (df_sal_NonNa.dropna() <= 250000)]
sb.distplot(df_sal_low)
# Here the value of the Converted Salary is less than or equal to 250000.
```

#### Here we draw the histogram graph with the help of a distplot() in which we enter the non NULL(Nan) of Converted Salary and  uses that value of Converted Salary which is less than or equals to 250000 which is stored in df_sal_low.
#### Here bins are selected by Default.

```python
# G
ax = sb.kdeplot(df_sal_NonNa)
sb.kdeplot(df_sal_low, ax=ax)
# Here the value of the df_sal_NonNa is greater in numbers than df_sal_low but both have non-null values. 
```

#### Here in this part we draw the KDE (kernel density estimation is a non-parametric way to estimate the probability density function (PDF) of a random variable. This function uses Gaussian kernels and includes automatic bandwidth determination ) plot of two df_sal_NonNa (non NULL values of Converted Salary ) and df_sal_low (Non NULL as well as less than or equal to 250000 of Converted Salary).
#### Here the probablity density function of the df_sal_low is less than that of df_sal_NonNa.
